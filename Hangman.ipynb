```python
# Importing the choice function from the random module at the beginning of the script
from random import choice

# Defining a Hangman class to encapsulate the game logic
class Hangman:
    # Constructor for the Hangman class
    def __init__(self, word_list, num_lives):
        # Randomly select a secret word from the word list and convert it to uppercase
        self.secret_word = choice(word_list).upper()
        # Initialize the number of lives the player has
        self.num_lives = num_lives
        # Calculate the number of unique letters in the secret word
        self.num_letters = len(set(self.secret_word))
        # Initialize a set to keep track of correctly guessed letters
        self.correct_letters = set()
        # Initialize a set to keep track of all guessed letters
        self.guesses = set()

    # Method to prompt the player to input a letter and process the guess
    def ask_for_input(self):
        guess = input("Enter a single letter: ").upper()
        # Validate the input and proceed only if it is a single alphabetical character
        if len(guess) == 1 and guess.isalpha():
            if guess in self.guesses:
                print("You already guessed that letter.")
            elif guess in self.secret_word:
                print("Good guess!")
                self.guesses.add(guess)
                self.correct_letters.add(guess)
                self.num_letters -= 1
            else:
                print("Sorry, that letter is not in the word.")
                self.guesses.add(guess)
                self.num_lives -= 1
        else:
            print("Invalid input. Please enter a single alphabetical character.")

    # Placeholder method for displaying the current game status
    def display_status(self):
        # This method should be implemented to display the game's current status
        # such as the hangman drawing and correctly guessed letters
        pass

# List of words used as potential secret words for the Hangman game
word_list = ['apple', 'banana', 'orange', 'grape', 'pear']

# Function to play the Hangman game
def play_game(word_list):
    # Set the initial number of lives
    num_lives = 5
    # Create a Hangman game instance
    game = Hangman(word_list, num_lives)

    # Main game loop
    while True:
        # Display the current status of the game
        game.display_status()
        
        # Check the state of the game and print messages accordingly
        if game.num_lives <= 0:
            print('You lost!')
            break
        elif game.num_letters > 0:
            game.ask_for_input()
        else:
            print('Congratulations. You won the game!')
            break

# Start the game by calling the play_game function with the word list
play_game(word_list)
```
